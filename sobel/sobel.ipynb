```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = plt.imread("signs_vehicles_xygrad.png")
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
abs_sobelx = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
abs_sobley = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))

#plt.imshow(abs_sobelx)
#plt.imshow(abs_sobley)

scaled_sobelx = np.uint8(255 * abs_sobelx/np.max(abs_sobelx))
scaled_sobely = np.uint8(255 * abs_sobley/np.max(abs_sobley))
#plt.imshow(scaled_sobelx)
#plt.imshow(scaled_sobely)

binary_outputx = np.zeros_like(scaled_sobelx)
binary_outputy = np.zeros_like(scaled_sobely)
#plt.imshow(binary_outputx)
#plt.imshow(binary_outputy)

binary_outputx[(scaled_sobelx >= 0) & (scaled_sobelx <= 100)] = 1
binary_outputy[(scaled_sobely >= 0) & (scaled_sobely <= 100)] = 1

#plt.imshow(binary_outputx)
#plt.imshow(binary_outputy)
```
